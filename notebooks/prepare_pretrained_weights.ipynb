```python
import torch
```

```python
pretrained_checkpoint = torch.load("exp/pretrained/model_best.pth")
pretrained_weights = pretrained_checkpoint["state_dict"]
```

```python
for key in pretrained_weights:
    print(key, pretrained_weights[key].shape)
```

# Remove last head layer

```python
last_head_keys = [
    key
    for key in pretrained_weights.keys()
    if "seg_head.3" in key
]

for key in last_head_keys:
    if "seg_head" in key:
        print(f"Remove: {key}: {pretrained_weights[key].shape}")
        del pretrained_weights[key]
```

```python
torch.save(pretrained_checkpoint, "exp/pretrained/pt-v2m2-keys.pth")
```

# Remove "module." prefix for single GPU training (not using DDP)

```python
prefix_removed_state_dict = {
    key.removeprefix("module."): pretrained_weights[key]
    for key in pretrained_weights
}

pretrained_checkpoint["state_dict"] = prefix_removed_state_dict
```

```python
torch.save(pretrained_checkpoint, "exp/pretrained/pt-v2m2-keys-modified.pth")
```
